```python
import matplotlib.pyplot as plt
from knetworks import knetworks, Network
import numpy as np
from generator import generateFromGenFile
import torch

%config InlineBackend.figure_format='retina'
device = torch.device("cuda")
```

```python
features = []
vectors = []
for f in range(10, 14):
    file = open("testdata/" + str(f) + ".txt", "r") 
    lines = file.readlines()
    features = sorted(str(lines[0]).replace(" ", "").upper().strip().split(","))
    vect = []
    for l in range(len(lines)-1):
        vec = lines[l+1].upper().split(",")
        vect.append([int(i) for i in vec])
    vectors.append(vect)
vectors = np.array(vectors)
data = vectors[:,:-4,:]
```

```python
knet = knetworks(2, data, len(features), device) # create a knetworks object
```

```python
knet.load("saves") # loading a savestate
```

```python
knet.fit(7) # fit the centroids to the data (using 7 datapoints for the mean)
```

```python
# Plotting the first 10 product from every user
# orange: training-set
# blue: the ground truth to be predicted by the model
f, axes = plt.subplots(4, 10, figsize=(30, 6))
count = 0
for i in range(4):
    for j in range(10):
        if count < len(features):
            axes[i][j].plot(np.array(vectors)[i][:,j])
            axes[i][j].plot(np.array(data)[i][:,j])
            count += 1
```

```python
knet.train(50,10) # weighted training on the CN's
```

```python
#knet.save("saves") # saving a savestate
```

```python
# red: ground truth
# green: prediction
# blue: rounded prediction

f, axes = plt.subplots(4, 10, figsize=(30, 6))
for i in range(4):
    prediction = np.array(np.squeeze(knet.predict(data[i], future=4)))
    for j in range(10):
        axes[i][j].plot(prediction[:,j], c='g', alpha=0.5)
        axes[i][j].plot(np.round(prediction[:,j]), alpha=0.5 , c='b')
        axes[i][j].plot(vectors[i][-4:,j], c='r', alpha=0.5)
```

```python
for user in range(4):
    prediction = np.array(np.squeeze(knet.predict(data[user], future=1))) # the prediction for user
    wrong_guesses = np.sum(np.absolute((np.absolute(np.round(prediction)) - vectors[user][-4,:]))) # the right guessed features for one receipt into the future
    print("["+ str(user) + "] Accuracy: " + str(len(features)-int(wrong_guesses)) + "/" + str(len(features)) + " (" + str((len(features)-int(wrong_guesses))/len(features)) + ")")
```
